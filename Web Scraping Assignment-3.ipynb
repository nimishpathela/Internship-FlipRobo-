```python
urls=[]
for page in range(0,3):
    url=amazon.find_elements(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for i in url:
        urls.append(i.get_attribute('href'))
nb=amazon.find_element(By.XPATH,'/html/body/div[1]/div[2]/div[1]/div[1]/div/span[1]/div[1]/div[64]/div/div/span/a[2]')
nb.click()
len(urls)
```

```python
brand=[]
for i in urls:
    amazon.get(i)
    time.sleep(1)
    try:
        bn=amazon.find_element(By.XPATH,'//td[@class="a-span9"]')
        brand.append(bn.text)
    except NoSuchElementException as e:
        brand.append('-')
```

```python
brand=[]
product_name=[]
price=[]
return_exchange=[]
expected_delivery=[]
availability=[]

for i in urls:
    amazon.get(i)
    time.sleep(1)
    try:
        bn=amazon.find_element(By.XPATH,'//td[@class="a-span9"]')
        brand.append(bn.text)
    except NoSuchElementException as e:
        brand.append('-')
        
    try:
        pn=amazon.find_element(By.XPATH,'//span[@class="a-size-large product-title-word-break"]')
        product_name.append(pn.text)
    except NoSuchElementException as e:
        product_name.append('-')

    try:
        pr=amazon.find_element(By.XPATH,'//span[@class="a-price-whole"]')
        price.append(pr.text)
    except NoSuchElementException as e:
        price.append('-')
    try:
        re=amazon.find_element(By.XPATH,'//span[@class="a-size-small a-color-link a-text-normal"]')
        return_exchange.append(re.text)
    except NoSuchElementException as e:
        return_exchange.append('-')
    try:
        ed=amazon.find_element(By.XPATH,'//span[@class="a-text-bold"]')
        expected_delivery.append(ed.text)
    except NoSuchElementException as e:
        expected_delivery.append('-')
    try:
        av=amazon.find_element(By.XPATH,'//span[@class="a-size-medium a-color-success"]')
        availability.append(av.text)
    except NoSuchElementException as e:
        availability.append('-')
```
