```python
wiki=webdriver.Chrome()
```

```python
wiki.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
wiki.maximize_window()
```

```python
search_bar=wiki.find_element(By.XPATH,'/html/body/div[1]/header/div[2]/div/div/div/form/div/div/div[1]/input')
search_bar.send_keys('most viewed videos on youtube')
search_button=wiki.find_element(By.XPATH,'/html/body/div[1]/header/div[2]/div/div/div/form/div/button')
search_button.click()
```

```python
Rank=[]
rank=wiki.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[1]')
for i in rank:
    Rank.append(i.text)

Name=[]
name=wiki.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[2]')
for i in name:
    Name.append(i.text)

Artist=[]
artist=wiki.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[3]")
for i in artist:
    Artist.append(i.text)
    
Views=[]
views=wiki.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[4]")
for i in views:
        Views.append(i.text)
UD=[]
upload_date=wiki.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[5]")
for i in upload_date:
        UD.append(i.text)
        
```

```python
len(rank),len(Name),len(Artist),len(Views),len(UD)
```

```python
Most_viewd_videos_on_YouTube=pd.DataFrame({"Rank":Rank, "Name":Name, "Artist":Artist, "Views":Views, "Updload Date":UD})
Most_viewd_videos_on_YouTube
```

Scrape the details most watched tv series of all time from imdb.com. Url = https://www.imdb.com/list/ls095964455/

B) Year span
C) Genre
D) Run time E) Ratings F) Votes

```python
imdb=webdriver.Chrome()
imdb.maximize_window()
imdb.get('https://www.imdb.com/list/ls095964455/')
```

```python
N=[]
names=imdb.find_elements(By.XPATH,'//div/h3/a')
for i in names:
    N.append(i.text)

YS=[]
yearspan=imdb.find_elements(By.XPATH,'//*[@class="lister-item-year text-muted unbold"]')
for i in yearspan:
    YS.append(i.text)
    
G=[]
genre=imdb.find_elements(By.XPATH,'//*[@class="genre"]')
for i in genre:
    G.append(i.text)

RT=[]
runtime=imdb.find_elements(By.XPATH,'//*[@class="runtime"]')
for i in runtime:
    RT.append(i.text)

R=[]
ratings=imdb.find_elements(By.XPATH,'//div[@class="ipl-rating-star small"]/span[@class="ipl-rating-star__rating"]')
for i in ratings[:100]:
    R.append(i.text)

V=[]
votes=imdb.find_elements(By.XPATH,'//p/span[@name="nv"]')
for i in votes:
    V.append(i.text)
```

```python
len(N),len(YS),len(G),len(RT),len(R),len(V)
```

```python
links=[]
link=uci.find_elements(By.XPATH,'//a[@class="link-hover link text-xl font-semibold"]')
for i in link:
    links.append(i.get_attribute('href'))
```

```python
DN=[]
dn=uci.find_elements(By.XPATH,'//a[@class="link-hover link text-xl font-semibold"]')
for i in dn:
    DN.append(i.text)
```

```python
DT=[]
T=[]
AT=[]
I=[]
A=[]
Y=[]
for i in links:
    uci.get(i)
    try:
        dt=uci.find_element(By.XPATH,'(//p[@class="text-md"])[1]')
        DT.append(dt.text)
    except NoSuchElementException:
        DT.append("-")
        
    try:
        t=uci.find_element(By.XPATH,'(//p[@class="text-md"])[3]')
        T.append(t.text)
    except NoSuchElementException:
        T.append("-")
        
    try:
        at=uci.find_element(By.XPATH,'(//p[@class="text-md"])[4]')
        AT.append(at.text)
    except NoSuchElementException:
        AT.append("-")
        
    try:
        instance=uci.find_element(By.XPATH,'(//p[@class="text-md"])[5]')
        I.append(instance.text)
    except NoSuchElementException:
        I.append("-")
        
    try:
        a=uci.find_element(By.XPATH,'(//p[@class="text-md"])[6]')
        A.append(a.text)
    except NoSuchElementException:
        A.append("-")
    
    try:
        y=uci.find_element(By.XPATH,'//h2[@class="text-primary-content"]')
        Y.append(y.text)
    except NoSuchElementException:
        Y.append("-")
            
```

```python
len(DN),len(DT),len(T),len(AT),len(I),len(A),len(Y)
```
